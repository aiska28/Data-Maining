```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("Advertising.csv")
data.head()
```

```python
data.info()
data.describe()
```

```python
x = data[['Advertising_Budget_USD']]
y = data['Product_Sales_Units']
```

```python
plt.scatter(x,y)
plt.xlabel("Advertising Budget (USD)")
plt.ylabel("Product Sales (Units)")
plt.title("Advertising Buget vs Product Sales")
plt.show()
```

```python
model = LinearRegression()
model.fit(x, y)

print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

```python
y_pred = model.predict(x)

plt.scatter(x, y)
plt.plot(x, y_pred)
plt.xlabel("Advertising Budget")
plt.ylabel("Product Sales")
plt.title("Linear Regression Result")
plt.show()
```

```python
prediksi = model.predict([[2500]])
print("Prediksi penjualan:", prediksi[0])
```
