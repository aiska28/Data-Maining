```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
# Load data Iris
iris = load_iris()
X_iris = iris.data
```

```python
data_kelas = pd.read_csv("Jobsheet_1.csv")
X_kelas = data_kelas.iloc[:, :-1]  # tanpa label
```

```python
wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X_iris)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 10), wcss)
plt.title("Elbow Method")
plt.xlabel("Jumlah Cluster")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
labels_iris = kmeans.fit_predict(X_iris)
```

```python
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=labels_iris, cmap='viridis')
plt.title("K-Means Iris")
plt.show()
```

```python
import pandas as pd

data_kelas = pd.read_csv("Jobsheet_1.csv")
X_kelas = data_kelas.iloc[:, :-1]  # tanpa label
```

```python
print(data_kelas.head())
print(data_kelas.info())
```

```python
X_kelas = data_kelas.select_dtypes(include=['int64', 'float64'])
```

```python
print(X_kelas.head())
print(X_kelas.dtypes)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X_kelas)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 10), wcss)
plt.title("Elbow Method Data Kelas")
plt.xlabel("Jumlah Cluster")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
labels_kelas = kmeans.fit_predict(X_kelas)
```

```python
plt.scatter(X_kelas.iloc[:, 0], X_kelas.iloc[:, 1], c=labels_kelas, cmap='viridis')
plt.title("K-Means Data Kelas")
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X_iris = iris.data

# Scaling
scaler = StandardScaler()
X_iris_scaled = scaler.fit_transform(X_iris)

dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_db_iris = dbscan.fit_predict(X_iris_scaled)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_iris[:, 0], X_iris[:, 1], c=labels_db_iris, cmap='viridis')
plt.title("DBSCAN Iris")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_kelas_scaled = scaler.fit_transform(X_kelas)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.8, min_samples=3)
labels_db_kelas = dbscan.fit_predict(X_kelas_scaled)
```

```python
plt.scatter(X_kelas.iloc[:, 0], X_kelas.iloc[:, 1], c=labels_db_kelas, cmap='viridis')
plt.title("DBSCAN Data Kelas")
plt.show()
```
